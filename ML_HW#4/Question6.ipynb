```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
dimentions = [5, 10, 100, 500, 1000, 10000, 1000000]

for dimention in dimentions:
    points = np.random.normal(size=(5, dimention))
    distances = []
    for i in range(len(points)):
        for j in range(i+1, len(points)):
            distances.append(np.linalg.norm(points[i]-points[j]))
    
    plt.bar(range(len(distances)), distances, facecolor='blue', alpha=0.5)
    plt.show()
```

```python
from sklearn.metrics import pairwise_distances

datapoin_nums = [5, 10, 100, 5000, 10000]

for datapoin_num in datapoin_nums:
    points = np.random.normal(size=(datapoin_num, 10000))
    X = pairwise_distances(points)
    distances = X[np.triu_indices(X.shape[0], 1)]
    
    plt.bar(range(len(distances)), distances, facecolor='blue', alpha=0.5)
    plt.show()
```
